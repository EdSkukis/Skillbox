````python
# Ваш код здесь
def binary_search(arr, k):
    i, left, right = 0, 0, len(arr) - 1
    while left <= right:
        i += 1
        mid = (left + right) // 2
        if k > arr[mid]:
            left = mid + 1 # Поиск в правой половине
        elif k < arr[mid]:
            right = mid - 1
        else:
            return [True, i]
    return [False, i]


numbers_list = [100, 200, 700, 1000, 1200]
key = 1000

result = binary_search(numbers_list, key)
print(*result)
````

*Пример 1:*

*Пример 2:*

Input:
```python
transactions = [100, 98, 1000, 2500, 299, 1898, 1989, 2001, 50, 10, 70]
```
Output:
```bash
[98, 100, 299, 1000, 1898, 1989, 50, 10, 70, 2001, 2500]
```

````python
# Первый пример для отладки
transactions = [100, 200, 90, 70, 120]

# Второй пример для отладки
# transactions = [100, 98, 1000, 2500, 299, 1898, 1989, 2001, 50, 10, 70]
````

````python
# Ваш код здесь
def move_max_to_end(arr, max_val):
    for i in range(len(arr)):
        if max_val == arr[i]:
            arr[i], arr[-1] = arr[-1], arr[i]
            return arr
            break

max_transaction = max(transactions)
transactions = move_max_to_end(transactions, max_transaction)
max_transaction = max(transactions[:-1])
transactions[-1], transactions[-2] = transactions[-2], transactions[-1]
transactions = move_max_to_end(transactions, max_transaction)
transactions[-1], transactions[-2] = transactions[-2], transactions[-1]
print(transactions)
````

**Способ №3 (выполните по желанию)**. Это самое эффективное и `элегантное` решение.

Существует логическая операция `XOR`, или исключающее «или». Логическая — значит, что она выполняется над битами числа в двоичной системе счисления. Берётся первый бит (1 или 0), затем второй, и с этими двумя битами выполняется  логическая операция, результатом которой будет определённый третий бит. Каждый бит принимает только два значения (1 или 0), поэтому для любой логической операции возможны всего четыре комбинации операндов и результата, которые в совокупности называются `таблицей истинности` этой логической операции. Таблица истинности логической операции `XOR`:

```bash
0 xor 0 = 0
1 xor 0 = 1
0 xor 1 = 1
1 xor 1 = 0
```

Результат `XOR` равен единице, только если операнды разные. Если операнды одинаковые, результат `XOR` равен нулю.

У этого свойства есть интересное следствие. Возьмём два одинаковых числа, например `5` и `5`, и запишем их в двоичной системе счисления (в столбик, одно число под другим).

```bash
101
101
```

Числа одинаковые, поэтому в двоичной форме они тоже выглядят одинаково. Теперь последовательно применим `XOR` к битам наших чисел: первый бит _поксорим_ с первым, второй — со вторым и так далее. Так как числа одинаковые, биты тоже будут одинаковыми, поэтому результат `XOR` будет равен нулю. Биты `одинаковых` чисел зануляют друг друга. Получается, что если одно число `поксорить` само на себя, то будет `0`: `5 xor 5 = 0`, `10 xor 10 = 0`, `42 xor 42 = 0`.

Далее `поксорим` `0` с каким-нибудь числом, например с 5. Для этого запишем числа в столбик в двоичной форме:

```bash
000
101
```

Все биты нуля ожидаемо нулевые, у пятёрки на некоторых позициях есть единицы. `XOR` устроен так, что для разных битов он даёт `1`, а для одинаковых — `0`. У нуля и любого другого числа одинаковые биты могут быть только нулевыми (`0 xor 0 = 0`), а разные — только если у другого числа на соответствующих позициях стоят единицы (`0 xor 1 = 1`). Получается, что нулевые биты другого числа нуль оставляет нулями, а единичные — единицами. Другими словами, нуль, `поксоренный` с любым другим числом, не изменяет это другое число: `0 xor 5 = 5`, `0 xor 10 = 10`, `0 xor 42 = 42`.

В Python есть специальный оператор, который выполняет `XOR` операндов, — это `^`.

```bash
42 ^ 42 = 0
0  ^ 42 = 42
42 ^ 0  = 42
```

Теперь вы знаете всё, чтобы эффективно решить задачу с поиском одинокого числа. Это задача «со звёздочкой», поэтому будет здорово, если у вас получится!

````python
numbers_list = [0, 1, 3, 4, 7, 98, 3, 1, 98, 7, 16, 16, 4, 0, 42]
````

````python
# Ваш код здесь
# Способ №2
numbers_list = [0, 1, 3, 4, 7, 98, 3, 1, 98, 7, 16, 16, 4, 0, 42]
count_numbers = [0] * 100
for i in range(len(numbers_list)):
    count_numbers[numbers_list[i]] += 1

for i in range(len(count_numbers)):
    if count_numbers[i] == 1:
        print(i)
        break
````

````python
def find_single_number(numbers_list):
    result = 0
    for num in numbers_list:
        result ^= num
    return result

numbers_list = [0, 1, 3, 4, 7, 98, 3, 1, 98, 7, 16, 16, 4, 0, 42]
single_number = find_single_number(numbers_list)
print(single_number)
````

***Задача 9. Сортировка и слияние файлов***

`Задача с собеседований` — _задачи с такой пометкой часто встречаются в практике собеседований для IT-специалистов_.

В оригинальной форме эта задача выглядит так: представьте себе, что вы работаете на компьютере с очень ограниченной памятью. На жёстком диске этого компьютера лежат два больших файла: `file1` и `file2`, в каждом файле — числа. Считать в память компьютера можно только один файл, но никак не оба одновременно. **Необходимо записать на жёсткий диск третий файл, который будет содержать числа из обоих файлов, отсортированные в порядке возрастания.**

Это старая задача из тех времён, когда компьютеры были большими и дорогими и постоянно приходилось экономить ресурсы. В эпоху больших данных эта задача снова обретает смысл.

Так как в памяти помещается только один файл, вы загружаете первый, сортируете (не важно как, хоть пузырьком) и отсортированным сохраняете на диск. Затем освобождаете память, загружаете в неё второй файл и делаете то же самое: сортируете и записываете на диск. Теперь на диске лежат два отсортированных файла. Пусть они выглядят так:

|file1 |file2 |
|------|------|
|  1   | 2    |
|  5   | 3    |
|  6   | 5    |
|  18  | 7    |
|  42  | 8    |
|  74  | 18   |
|  80  | 19   |
|  87  | 20   |
|  91  | 43   |
|  99  | 50   |
|  100 | 81   |
|      | 93   |
|      | 101  |
|      | 102  |
|      | 103  |

По отдельности каждый файл отсортирован, но вам нужно объединить эти два файла (или, как ещё говорят, _слить_) в один отсортированный файл. Считать их в память, соединить и отсортировать не получится, потому что одновременно они в памяти не поместятся. Поэтому нужен алгоритм, который позволит записывать объединённый отсортированный файл, не считывая в память два оригинальных файла. Формально такой алгоритм называется [сортировка слиянием](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC#:~:text=%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0%20%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC%20(%D0%B0%D0%BD%D0%B3%D0%BB.,%D0%BF%D1%80%D0%B8%D0%BD%D1%86%D0%B8%D0%BF%D0%B0%20%C2%AB%D1%80%D0%B0%D0%B7%D0%B4%D0%B5%D0%BB%D1%8F%D0%B9%20%D0%B8%20%D0%B2%D0%BB%D0%B0%D1%81%D1%82%D0%B2%D1%83%D0%B9%C2%BB.).

Вам потребуется только часть этого алгоритма, потому что исходные файлы уже отсортированы. Остаётся только слить их в единый файл, который тоже будет отсортирован. Разберём алгоритм `слияния` по шагам:

1. Берём `первые` элементы каждого файла и сравниваем их между собой.
1. `1` меньше, чем `2`, поэтому сохраняем `1` в файл `result` (будет [1]).
1. Берём `второй` элемент из `file1` и сравниваем его с `первым элементом` из `file2`.
1. `5` больше, чем `2`, поэтому сохраняем `2` в файл `result` (будет [1, 2]).
1. Берём `второй` элемент из `file2` и сравниваем с ним `второй` элемент из `file1`.
1. `5` больше, чем `3`, поэтому сохраняем `3` в файл `result` (будет [1, 2, 3]).
1. Берём `третий` элемент из `file2` и сравниваем с ним `второй` элемент из `file1`.

***Задача 10. Скользящее среднее***

*Скользящее среднее* — специальная функция, которая в каждой точке равна среднему арифметическому значений исходной функции за предыдущий период. Скользящие средние используют для _сглаживания_ краткосрочных колебаний в данных и выделения в них основных трендов.

Пусть дан список `numbers_list = [2, 12, 7, 8, 0]`. Его элементы можно рассматривать как значения некоторой функции. Возьмём первые три элемента этого списка и посчитаем их арифметическое среднее:

[**2**, **12**, **7**, 8, 0] —> (2 + 12 + 7) / 3 = 7

Теперь берём второй, третий и четвёртый элементы списка и считаем их среднее:

[2, **12**, **7**, **8**, 0] —> (12 + 7 + 8) / 3 = 9

Наконец, берём третий, четвёртый и пятый элементы списка и считаем их среднее:

[2, 12, **7**, **8**, **0**] —> (7 + 8 + 0) / 3 = 5

В таком случае скользящим средним списка с окном в три элемента будет список `[7, 9, 5]`.

Обратите внимание, что окно размером в три элемента как бы `скользит` по списку, сдвигаясь вправо на один элемент. На каждом шаге этого окна вычисляется среднее арифметическое входящих в него элементов, затем добавляется в новый список, который и будет результатом скользящего среднего с заданным окном для исходного списка.
